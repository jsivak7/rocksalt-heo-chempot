## Plotting AO Chemical Potential Diagrams 
- Now includes all Mg + 10 3d TMs

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.dpi"] = 400 # should be 400
plt.rcParams["figure.figsize"] = (1, 4)
plt.rcParams["font.size"] = 10
plt.rcParams["font.serif"] = "Helvetica"
plt.rcParams["axes.linewidth"] = 1.65
plt.rcParams["xtick.direction"] = "in"
plt.rcParams["ytick.direction"] = "in"
plt.rcParams["xtick.major.width"] = 1.65
plt.rcParams["ytick.major.width"] = 1.65
plt.rcParams["font.weight"] = "bold"
```

```python
cation = "Sc"

A_O = [
    [-50, -9.9218, 0,          0], # x
    [0,   0,       -6.6146,    -50], # y
    ]

plt.plot(A_O[0], A_O[1], marker="s", linewidth=2, color="k")
#plt.plot(A_O[0][1:3], A_O[1][1:3], marker="s", linewidth=2, color="tab:green")

plt.xticks([-8, -4, 0])

plt.xlim(A_O[0][1]-2, A_O[0][-1]+2)
plt.ylim(-7.75, 0.75)

plt.xlabel(r"$_{\rm \bf A}$ - $_{\rm \bf A}^{\rm \bf o}$ (eV)", weight="bold")
plt.ylabel(r"$_{\rm \bf O}$ - $_{\rm \bf O}^{\rm \bf o}$ (eV)", weight="bold")

plt.title(f"{cation} - O", weight="bold")

plt.show()
```
