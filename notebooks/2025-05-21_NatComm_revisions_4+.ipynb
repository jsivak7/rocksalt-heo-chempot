```python
import matplotlib.pyplot as plt

plt.rcParams["figure.dpi"] = 400
plt.rcParams["figure.figsize"] = (1, 4)
plt.rcParams["font.size"] = 10
plt.rcParams["font.serif"] = "Helvetica"
plt.rcParams["axes.linewidth"] = 1.65
plt.rcParams["xtick.direction"] = "in"
plt.rcParams["ytick.direction"] = "in"
plt.rcParams["xtick.major.width"] = 1.65
plt.rcParams["ytick.major.width"] = 1.65
plt.rcParams["font.weight"] = "bold"
```

```python
Ce_O = [
    [-50, -11.785389,   -4.329403,  -2.588021,  -0.934633,  -0.555993,  0.000000, 0], # x
    [0,   0,            -3.727993,  -4.695427,  -5.659904,  -5.912330,  -6.468324, -50], # y
    ]

plt.fill_between(
    [Ce_O[0][1] - 2, Ce_O[0][-1] + 2],
    y1=Ce_O[1][1],
    y2=Ce_O[1][2],
    color="tab:green",
    alpha=0.2,
    zorder=1,
    linewidth=0,
)

plt.plot(Ce_O[0], Ce_O[1], marker="s", linewidth=2, color="k")

plt.xticks([-8, 0])

plt.xlim(Ce_O[0][1]-2, Ce_O[0][-1]+2)
plt.ylim(-7.75, 0.75)

plt.xlabel(r"μ$_{\rm \bf A}$ - μ$_{\rm \bf A}^{\rm \bf o}$ (eV)", weight="bold")
plt.ylabel(r"μ$_{\rm \bf O}$ - μ$_{\rm \bf O}^{\rm \bf o}$ (eV)", weight="bold")

plt.title(r"Ce$^{\rm \bf 4.0\!+}$", weight="bold")

plt.show()
```
