## Plotting Chemical Potential Diagrams for NatComm Review

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.dpi"] = 200
plt.rcParams["figure.figsize"] = (1, 4)
plt.rcParams["font.size"] = 10
plt.rcParams["font.serif"] = "Arial"
plt.rcParams["axes.linewidth"] = 1.00
plt.rcParams["xtick.direction"] = "in"
plt.rcParams["ytick.direction"] = "in"
plt.rcParams["xtick.major.width"] = 1.00
plt.rcParams["ytick.major.width"] = 1.00
plt.rcParams["font.weight"] = "normal"
```

```python
# 3+
Ce_O = [
    [-50, -6.1068, 0      , 0], # x
    [0,   0,       -6.1068, -50], # y
    ]

plt.fill_between(
    [Ce_O[0][1] - 1, Ce_O[0][-1] + 1],
    y1=Ce_O[1][1],
    y2=Ce_O[1][2],
    color="tab:green",
    alpha=0.2,
    zorder=1,
    linewidth=0,
)

plt.plot(Ce_O[0], Ce_O[1], marker="s", linewidth=1.5, color="k")
#plt.plot(Mg_O[0][1:3], Mg_O[1][1:3], marker="s", linewidth=1.5, color="tab:green")

plt.xticks([-4, 0])

plt.xlim(Ce_O[0][1]-1, Ce_O[0][-1]+1)
plt.ylim(-7.75, 0.75)

plt.xlabel(r"μ$_{\rm A}$ - μ$_{\rm A}^{\rm o}$ (eV)", weight="normal")
plt.ylabel(r"μ$_{\rm O}$ - μ$_{\rm O}^{\rm o}$ (eV)", weight="normal")

plt.title("Ce3+ - O", weight="bold")

plt.show()
```
