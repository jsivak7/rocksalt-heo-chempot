```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = [7, 5.5]
d = pd.read_csv('./overlaps.csv')
d_solo = pd.read_csv('./AO-Oxygen-chempot-stablerange-MP-GGA_GGAU_R2SCAN.csv')
```

```python
map = [
[-d_solo.MgO[2],    np.nan,             np.nan,             np.nan,             np.nan,             np.nan,             np.nan,             np.nan],    # Mg
[d.overlap_eV[0],   -d_solo.CaO[2],     np.nan,             np.nan,             np.nan,             np.nan,             np.nan,             np.nan],    # Ca
[d.overlap_eV[1],   d.overlap_eV[7],    -d_solo.MnO[2],     np.nan,             np.nan,             np.nan,             np.nan,             np.nan],    # Mn
[d.overlap_eV[2],   d.overlap_eV[8],    d.overlap_eV[13],   -d_solo.FeO[2],     np.nan,             np.nan,             np.nan,             np.nan],    # Fe
[d.overlap_eV[3],   d.overlap_eV[9],    d.overlap_eV[14],   d.overlap_eV[18],   -d_solo.CoO[2],     np.nan,             np.nan,             np.nan],    # Co
[d.overlap_eV[4],   d.overlap_eV[10],   d.overlap_eV[15],   d.overlap_eV[19],   d.overlap_eV[22],   -d_solo.NiO[2],     np.nan,             np.nan],    # Ni
[d.overlap_eV[5],   d.overlap_eV[11],   d.overlap_eV[16],   d.overlap_eV[20],   d.overlap_eV[23],   d.overlap_eV[25],   -d_solo.CuO[2],     np.nan],    # Cu
[d.overlap_eV[6],   d.overlap_eV[12],   d.overlap_eV[17],   d.overlap_eV[21],   d.overlap_eV[24],   d.overlap_eV[26],   d.overlap_eV[27],   -d_solo.ZnO[2]],    # Zn
#   Mg              Ca                  Mn                  Fe                  Co                  Ni                  Cu                  Zn
]
```

```python
elements = ['Mg', 'Ca', 'Mn', 'Fe', 'Co', 'Ni', 'Cu', 'Zn']
overlap_df = pd.DataFrame(map, index = elements, columns = elements)
```

```python
ax = sns.heatmap(overlap_df, 
    linewidths=1,
    cmap = sns.color_palette("seismic", as_cmap=True), 
    cbar_kws={'label': 'Stable $\mu_{O}$ Window (eV)'},
    annot = True,
    fmt = '.2f',
    center = 0,
    vmin = -7,
    vmax = 7
    )
```
